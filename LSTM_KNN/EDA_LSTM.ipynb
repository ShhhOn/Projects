```python
from keras.models import Sequential
import numpy as np
from keras.layers import LSTM, Input, Dropout
from keras.layers import Dense
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from keras.models import Model
```

```python
#create single variable time series to practice LSTM.  Use '2_LT_002_PV'as it has an attack 10/10/17 10:55 to 11:24
df_test = df_WADIi_a[['Time_Stamp', '2_LT_002_PV']]
```

```python
df_test.plot('Time_Stamp', '2_LT_002_PV')
```

```python
train_dat, test_dat = df_test.loc[df_test['Time_Stamp'] > '2017/10/11 06:00:00'], df_test.loc[df_test['Time_Stamp'] <= '2017/10/11 06:00:00']
```

```python
train_dat.plot('Time_Stamp', '2_LT_002_PV')
```

```python
test_dat.plot('Time_Stamp', '2_LT_002_PV')
```

```python
scaler = StandardScaler()
scaler = scaler.fit(train_dat[['2_LT_002_PV']])
```

```python
train_dat['2_LT_002_PV'] = scaler.transform(train_dat[['2_LT_002_PV']])
test_dat['2_LT_002_PV'] = scaler.transform(test_dat[['2_LT_002_PV']])
```

```python
train_dat.plot('Time_Stamp', '2_LT_002_PV')
```

```python
seq_size = 1200
```

```python
def to_sequences(x, y, seq_size = 1):
    x_values = []
    y_values = []

    for i in range(len(x)-seq_size):
        x_values.append(x.iloc[i:(i+seq_size)].values)
        y_values.append(y.iloc[i+seq_size])

    return np.array(x_values), np.array(y_values)
```

```python
trainX, trainY = to_sequences(train_dat[['2_LT_002_PV']], train_dat[['2_LT_002_PV']], seq_size)
testX, testY = to_sequences(test_dat[['2_LT_002_PV']], test_dat[['2_LT_002_PV']], seq_size)
```

```python
trainX
```

```python
model = Sequential()
model.add(LSTM(4, input_shape=(trainX.shape[1], trainX.shape[2])))
model.add(Dropout(rate=0.2))

```

```python
model.add(RepeatVector(trainX.shape[1]))

```

```python
model.add(LSTM(4, return_sequences=True))
model.add(Dropout(rate=0.2))
model.add(TimeDistributed(Dense(trainX.shape[2])))
model.compile(optimizer = 'adam',loss = 'mae')
model.summary
```

```python
history = model.fit(trainX, trainY, epochs=1, batch_size=120, validation_split=0.1, verbose=True)
```

```python
trainPredict = model.predict(trainX)
trainMAE = np.mean(np.abs(trainPredict - trainX), axis = 1)

```

```python
plt.hist(trainMAE, bins=30)
```

```python
testPredict = model.predict(testX)
testMAE = np.mean(np.abs(testPredict - testX), axis =1)
plt.hist(testMAE, bins=30)
```

```python
max_trainMAE = 2.2
```

```python
anomaly_df = pd.DataFrame(test_dat[seq_size:])
anomaly_df['testMAE'] = testMAE
anomaly_df['max_trainMAE'] = max_trainMAE
anomaly_df['anomaly'] = anomaly_df['testMAE'] > anomaly_df['max_trainMAE']
```
