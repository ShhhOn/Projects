```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from datetime import datetime, date

pd.set_option('max_columns', None)

%matplotlib inline
sns.set_style("whitegrid")
plt.style.use("fivethirtyeight")
```

```python
df23 = pd.read_csv("heart_disease_conditions_23M_20220319.csv")
df23.head()
```

```python
def age(born):
    # born = datetime.strptime(born, "%d/%m/%Y").date()
    #today = date.today()
    today = datetime.strptime('2010-06-30', '%Y-%m-%d')
    return today.year - born.year - ((today.month,
                                      today.day) < (born.month,
                                                    born.day))
```

```python
cols=["year_of_birth", "month_of_birth", "day_of_birth"]
df23['dob'] = df23[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
df23['dob']=pd.to_datetime(df23['dob'])
df23['age'] = df23['dob'].apply(age)
df23.head()
```

```python
df23.loc[df23['person_id'] == 34]
```

```python
#-65, 65-74, 75-84, 85+?
def age_group(age):
    if age < 65:
        return "-65"
    elif age>=65 and age<=74:
        return "65-74"
    elif age>=75 and age<=84:
        return "75-84"
    elif age>=85:
        return "85+"
    else:
        return "-65"  
    
age_group(10)
```

```python
df23['age_group'] = df23['age'].apply(age_group)
df23.head()
```
