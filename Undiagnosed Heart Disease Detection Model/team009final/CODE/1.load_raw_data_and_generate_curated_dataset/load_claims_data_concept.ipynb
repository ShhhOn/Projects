```
from pyspark.sql.types import *
 
schema = StructType([
  # Define a StructField for each field
  StructField('concept_id', IntegerType()),
  StructField('concept_name', StringType()),
  StructField('domain_id', StringType()),
  StructField('vocabulary_id', StringType()),
  StructField('concept_class_id', StringType()),
  StructField('standard_concept', StringType()) , 
    StructField('concept_code', StringType()),
    StructField('valid_start_date', DateType()),
    StructField('valid_end_date', DateType()),
    StructField('invalid_reason', StringType())
])

# df = spark.read.format('csv') \
# .schema(schema) \
# .option("header", "true") \
# .option("delimiter", "\t") \
# .load('/FileStore/pankaj/gatech/CONCEPT_csv.gz')

df = spark.read.format('csv').options(header='true', inferSchema='true', delimiter='\t').load('/FileStore/pankaj/gatech/CONCEPT_csv.gz')

df.printSchema()    
# df = df.toDF('concept_id','concept_name','domain_id','vocabulary_id','concept_class_id','standard_concept','concept_code','valid_start_date','valid_end_date','invalid_reason')
spark.sql("DROP TABLE IF EXISTS pankaj_gt.concept") 
df.write.saveAsTable("pankaj_gt.concept")
df.show()
```
